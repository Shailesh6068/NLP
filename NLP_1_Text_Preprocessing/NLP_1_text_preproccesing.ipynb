# Remove URL

```python
import re

url_pattern = re.compile(r'https?://\S+|www\.\S+')

def remove_url(text):
    return url_pattern.sub('', text)
```

```python
text1 = 'Check out my notebook https://www.kaggle.com/campusx/notebook8223fc1abb'
text2 = 'Check out my notebook http://www.kaggle.com/campusx/notebook8223fc1abb'
text3 = 'Google search here www.google.com'
text4 = 'For notebook click https://www.kaggle.com/campusx/notebook8223fc1abb to search check www.google.com'

print(remove_url(text1))
print(remove_url(text2))
print(remove_url(text3))
print(remove_url(text4))
```

# Short Word / Chat Word Handle

```python
chat_words = {
    "BRB": "Be Right Back",
    "LOL": "Laughing Out Loud",
    "OMG": "Oh My God",
    "IDK": "I Don't Know",
    "BTW": "By The Way",
    "ASAP": "As Soon As Possible",
    "FYI": "For Your Information",
    "IMHO": "In My Honest Opinion",
    "TTYL": "Talk To You Later",
    "GM": "Good Morning",
    "GN": "Good Night",
    "THX": "Thanks",
    "PLZ": "Please",
    "U": "You",
    "UR": "Your",
    "BTW": "By The Way",
    "B4": "Before",
    "CU": "See You"
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
text = "GM bro LOL I will BRB IDK BTW"
print(chat_conversion(text))
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word.lower() not in stop_words:
            new_text.append(word)
    return " ".join(new_text)
```

```python
text = "This is a simple example to understand the removal of stop words in NLP"
print(remove_stopwords(text))
```

# Handling Emotions

```python
import re

def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map
        u"\U0001F1E0-\U0001F1FF"  # flags
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub('', text)
```

```python
text = "I love Python üòçüî•"
print(remove_emoji(text))
```

# Stemming 

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem_words(text):
    return " ".join(ps.stem(word) for word in text.split())

```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

```python
text = "probably my alltime favorite movie a story of selflessness sacrifice and dedication to a noble cause but its not preachy or boring it just never gets old"
print(stem_words(text))
```

# Lematization

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string

# Download required resources (run once)
# nltk.download('punkt')
# nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = word_tokenize(text)
    lemmas = [
        lemmatizer.lemmatize(word, pos='v')
        for word in words
        if word not in string.punctuation
    ]
    return " ".join(lemmas)

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
print(lemmatize_text(sentence))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string

# Download once if not already done
# nltk.download('punkt')
# nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."

# Tokenize sentence
words = word_tokenize(sentence)

# Remove punctuation
words = [word for word in words if word not in string.punctuation]

# Print header
print("{0:20}{1:20}".format("Word", "Lemma"))

# Lemmatize each word (verb POS for better result)
for word in words:
    print("{0:20}{1:20}".format(word, lemmatizer.lemmatize(word, pos='v')))
```
